# Import data

```python
import pymysql
import pandas as pd
import numpy as np

conn = pymysql.connect("localhost",user="akhil",passwd="akhil",db="enron")
cur = conn.cursor()
# cur.execute("show columns from employeelist")
```

```python
def createDFFromCursor(cursor):
    columns = [d[0] for d in cursor.description]
    cDict = {col:[] for col in columns}
    
    for row in cursor:
        for i in range(len(row)): cDict[columns[i]].append(row[i])
    
    df = pd.DataFrame(cDict)
#     df.columns = cDict.keys()
    
    return df
```

```python
cur.execute("select m.mid mid \
                    ,m.date SentDate \
                    ,m.sender SenderEmail \
                    ,es.eid SenderEID \
                    ,es.firstName SenderFirstName \
                    ,es.lastName SenderLastName \
                    ,r.rid rid \
                    ,r.dater ReceivedDate \
                    ,r.rvalue ReceiverEmail \
                    ,r.rtype ReceiverType \
                    ,er.eid ReceiverEID \
                    ,er.firstName ReceiverFirstName\
                    ,er.lastName ReceiverLastName \
            from message m inner join recipientinfo r on (r.mid = m.mid) \
            inner join employeelist es on (es.Email_id = m.sender) \
            inner join employeelist er on (er.Email_id = r.rvalue) \
            where m.sender like '%@enron.com' and r.rvalue like '%@enron.com' \
            order by m.date")

data = createDFFromCursor(cur)
data = data.sort("SentDate")
# for row in cur:
#     print row
# print cur.execute("select m.sender,m.date,r.rvalue \
#             from message m inner join recipientinfo r on (r.mid = m.mid and r.) \
#             where m.sender like '%@enron.com' and r.rvalue like '%@enron.com'")
```
